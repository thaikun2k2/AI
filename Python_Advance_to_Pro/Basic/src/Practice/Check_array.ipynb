cho 1 mảng cho trc kiểm tra xem có phải là mảng tăng dần hay không

eg:
input: [1, 2, 3, 4, 5]
output: True


input: [1, 3, 2, 4, 5]
output: False

```python
arr1 = [1, 2, 3, 4, 5]
arr2 = [1, 3, 2, 4, 5]


def check_arrays(arrr):
    for i in range(len(arrr) - 1):
        if arrr[i] >= arrr[i + 1]:
            return False
    return True

print(check_arrays(arr1))  # Output: True

print(check_arrays(arr2))  # Output: False  
```
