đổi vị trí phần tử đầu và phần tử cuối trong list
input: [12, 35, 9, 56, 24]
output: [24, 35, 9, 56, 12]

```python
list1 = [12, 35, 9, 56, 24]

def swapList(new_list1):
    a = len(new_list1)
    temp = new_list1[0]
    new_list1[0] = new_list1[a-1]
    new_list1[a-1] = temp
    return new_list1

print(swapList(list1))
```
