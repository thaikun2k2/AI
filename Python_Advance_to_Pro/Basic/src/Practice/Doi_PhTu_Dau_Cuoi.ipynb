đổi vị trí phần tử đầu và phần tử cuối trong list
input: [12, 35, 9, 56, 24]
output: [24, 35, 9, 56, 12]

```python

##Cách 1:
list1 = [12, 35, 9, 56, 24]

def swapList(new_list1):
    a = len(new_list1)
    temp = new_list1[0]
    new_list1[0] = new_list1[a-1]
    new_list1[a-1] = temp
    return new_list1

print(swapList(list1))
```

```python
##Cách 2:
list1 = [12, 35, 9, 56, 24]

def swapList(new_list1):

    temp = new_list1[0]
    new_list1[0] = new_list1[-1]
    new_list1[-1] = temp
    return new_list1

print(swapList(list1))
```

```python
##Cách 3:
list1 = [12, 35, 9, 56, 24]

def swapList(new_list1):

    temp = new_list1[0], new_list1[-1]
    new_list1[-1], new_list1[0] = temp
    return new_list1

print(swapList(list1))
```

```python
##Cách 4:
list1 = [12, 35, 9, 56, 24]

def swapList(new_list1):

    new_list1[0], new_list1[-1] = new_list1[-1], new_list1[0]
    return new_list1

print(swapList(list1))
```

```python
##Cách 5:
list1 = [12, 35, 9, 56, 24]

def swapList(new_list1):

    a, *b, c = new_list1
    new_list1 = [c, *b, a]
    return new_list1

print(swapList(list1))
```

```python
##Cách 6:
list1 = [12, 35, 9, 56, 24]

def swapList(new_list1):

    a, *b, c = new_list1
    new_list1[0] = c
    new_list1[-1] = a
    return new_list1

print(swapList(list1))
```
