Bài toán and Dataset

```python
import csv
import random
import operator
def load_Dataset(tenFile, split, trainingSet=[], testSet=[]):
    with open(tenFile, "r") as tep:
        lines = csv.reader(tep)
        dsDuLieu = list(lines)
        for x in range(len(dsDuLieu)):
            for y in range(4):
                dsDuLieu[x][y] = float(dsDuLieu[x][y])
            if random.random() < split:
                trainingSet.append(dsDuLieu[x])
            else:
                testSet.append(dsDuLieu[x])
```

Độ đo khoảng cách

Kc = sprt(((p1 - q1)^2 )*((p2 - q2)^2)*...*((pn - qn)^2))

```python
import math
def euclideanDistance(duKien1, duKien2, n):
    kc = 0
    for x in range(n):
        kc += pow((duKien1[x] - duKien2[x]), 2)
    return math.sqrt(kc)
```

```python
import operator
def getNeighbors(trainingSet, motDuKien, k):
    dsKhoangcach = []     # tạo 1 list
    n = len(motDuKien) - 1
    for x in range(len(trainingSet)):
        kc = euclideanDistance(motDuKien, trainingSet[x], n)
        dsKhoangcach.append((trainingSet[x], kc))
    dsKhoangcach.sort(key=operator.itemgetter(1))
    neighbors = []
    for x in range(k):
        neighbors.append(dsKhoangcach[x][0])
    return neighbors
```

Gán nhãn

```python
import operator

def prediction(neighbors):
    classVotes = {}  # tạo 1 dictionary
    for x in range(len(neighbors)):
        response = neighbors[x][-1]
        if response in classVotes:
            classVotes[response] += 1
        else:
            classVotes[response] = 1
    sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)
    return sortedVotes[0][0]
```

```python
neighbors = [
    [7.7, 3.8, 6.7, 2.2, 'Iris-virginica'],
    [7.6, 3.0, 6.6, 2.1, 'Iris-versicolor'],
    [7.7, 3.0, 6.1, 2.3, 'Iris-virginica'],
    [5.0, 3.4, 1.5, 0.2, 'Iris-setosa']
]

response = prediction(neighbors)
print(response)
```

Độ chính xác

```python
def getAccuracy(testSet, predictions):
    correct = 0
    for x in range(len(testSet)):
        if testSet[x][-1] == predictions[x]:
            correct += 1
    return (correct/float(len(testSet))) * 100.0
```

```python
test = [
    [7.7, 3.8, 6.7, 2.2, 'Iris-virginica'], 
    [7.6, 3.0, 6.6, 2.1, 'Iris-versicolor'], 
    [7.7, 3.0, 6.1, 2.3, 'Iris-virginica']
]

predictions = ['Iris-virginica', 'Iris-virginica', 'Iris-virginica']
doChinhXac = getAccuracy(test, predictions)
print(doChinhXac)
```

```python

def main():
    trainingSet = []
    testSet = []
    split = 0.75
    load_Dataset('C:/CNTT/AI/Machine_Learning/Practice/data/iris.data.csv', split, trainingSet, testSet)
    #load_Dataset('drive/My Drive/Colab Notebooks/data/iris.data.csv', split, trainingSet, testSet)
    print('Train set: ', repr(len(trainingSet)))
    print('Test set: ', repr(len(testSet)))

    # List để lưu trữ các dự đoán
    predictions = []
    k = 5
    for x in range(len(testSet)):
        neighbors = getNeighbors(trainingSet, testSet[x], k)
        result = prediction(neighbors)
        predictions.append(result)
        print('>du doan =', repr(result), ', thuc te =', repr(testSet[x][-1]))
    accuracy = getAccuracy(testSet, predictions)
    print('Accuracy:', repr(accuracy), '%')

# Gọi hàm main()
main()
```
