Tách dữ liệu

```python
def divide_set(rows, column, value):
  split_function = None

  if (isinstance(value, int) or isinstance(value, float)):
    split_function = lambda row: row[column] >= value
  else:
    split_function = lambda row: row[column] == value

  set1 = [row for row in rows if split_function(row)]
  set2 = [row for row in rows if not split_function(row)]
  return (set1, set2)
```

```python
# result cho 2 danh sách ứng vs 2 tập ở cột số '2'ứng vs 2 nút con trong cây phân tách
print(divide_set(my_data, 2, 'no'))
```

```python
# result cho 2 danh sách được phân tách ở cột số 3 và phân tách ở giá trị 20
print(divide_set(my_data, 3, 20))
```

```python
# result cho 2 danh sách, value = 'slashdot', dùng cột 0 để phân tách List
print (divide_set(my_data, 0, 'slashdot'))
```

**Entropy**

là đánh giá độ không chắc chắn của lượng thông tin nào đó

```python
#tính số lần xuất hiện của mỗi lớp Ci trong tập S
# tìm  Ci
def unique_counts (rows):
  results = {}
  for row in rows:
    r= row[len(row) -1]           
    if r not in results: 
      results[r] = 0
    results[r] += 1
  return results
```

```python
print(unique_counts(my_data))
```

pi = Ci/S

H(S) = -P1*log2(P1)- ... - Pn*log2(Pn)

```python
def entropy(rows):
    from math import log
    log2 = lambda x:log(x)/log(2)
    results = unique_counts (rows)

    # ent =H(S)
    ent = 0.0
    for r in results.keys():
      p = float (results[r]) / len(rows)
      ent = ent-p*log2(p)
    return ent
```

```python
print(entropy(my_data))
```

```python
set1, set2 = divide_set(my_data,3,20)
print(entropy(set1), entropy(set2))
```

```python
set3, set4 = divide_set(my_data, 0, 'slashdot')
print(entropy(set3), entropy(set4))
```

**Gini**

Gini = p(i) * p(j)

sum i != j

```python
def gini_impurity(rows):
    total=len(rows)
    counts=unique_counts(rows)
    imp=0
    for k1 in counts:
      p1 = float(counts[k1])/total
      for k2 in counts:
        if k1 ==k2: continue
        p2 = float(counts[k2])/total
        imp += p1*p2
    return imp
```

```python
print(gini_impurity(my_data))
```

**Dựng cây quyết định**

```python
class decision_node:
    def __init__(self, col=-1, value=None, results=None, tb=None, fb=None):
        self.col = col
        self.value = value
        self.results = results
        self.tb = tb
        self.fb = fb
```

```python
def build_tree(rows, scoref=entropy):        # or scoref=gini_impurity

    # rows là tập dữ liệu tại 1 nút muốn tách
    if len(rows) == 0: return decision_node()
    current_score = scoref(rows)

    # Khởi tạo biến để lưu vết
    best_gain = 0.0
    best_criteria = None
    best_sets = None

    # Tính số đặc trưng của rows. Không tính cột cuối cùng
    column_count = len(rows[0]) - 1
    #for col in range(0, column_count):
    for col in range(0, column_count):
    # Tạo ra List all các giá trị khác nhau có thể của cột data
        global column_values
        column_values = {}
    
        for row in rows:
            column_values[row[col]] = 1

        # Chia data ra 2 tập = cách duyệt qua mọi giá trị trg cột
        for value in column_values.keys():
            (set1, set2) = divide_set(rows, col, value)
        
            # 2 nhánh con, Tính IG
            p = float(len(set1)) / len(rows)
            gain = current_score - p * scoref(set1) - (1 - p) * scoref(set2)
            
            if gain > best_gain and len(set1) > 0 and len(set2) > 0:
                best_gain = gain
                best_criteria = (col, value)
                best_sets = (set1, set2)
        
    # Tạo ra nhánh con
    if best_gain > 0:
        trueBranch = build_tree(best_sets[0])
        falseBranch = build_tree(best_sets[1])
        return decision_node(col = best_criteria[0], value = best_criteria[1], tb = trueBranch, fb = falseBranch)
    else:
        return decision_node(results = unique_counts(rows))
        
```

```python
def classify(observation, tree):
    if tree.results != None:
        return tree.results
    else:
        v = observation[tree.col]
        branch = None
        if isinstance(v, int) or isinstance(v, float):
            if v >= tree.value: 
                branch = tree.tb
            else: 
                branch = tree.fb
        else:
            if v == tree.value: 
                branch = tree.tb
            else: 
                branch = tree.fb
        return classify(observation, branch)
```
